```python
import time
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

import warnings
warnings.filterwarnings('ignore')

import re
import random
from sys import stdout
from datetime import datetime
```

## 기본 링크 다운

#### 홈페이지에서 카테고리별 데이터 다운

```python
def base(url):
    options = webdriver.ChromeOptions()
    options.add_argument("disable-gpu")
    options.add_argument("lang=ko_KR")
    #options.add_argument("headless")

    driver = webdriver.Chrome('C:\\Users\ ',options=options) # 크롬드라이버 경로
    driver.get(url)

    menu = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div/header/div')

    ActionChains(driver).move_to_element(menu).perform()
    time.sleep(1)

    category_url = []
    category = []
    for i in range(2, 16):
        try:
            hidden1 = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div/header/div/div/ul[1]/li[%d]/a' %i).get_attribute("href")
        except:
            hidden1 = None

        try:
            name1 = driver.find_element(by=By.XPATH, value = '/html/body/div[3]/div/header/div/div/ul[1]/li[%d]/a' %i).text
        except:
            name1 = None

        category_url.append(hidden1)
        category.append(name1)
    driver.close()


```

## 제품 url

#### 상위 60개, 1page

```python
def coupang_product(cat, url):
    
    options = webdriver.ChromeOptions()
    options.add_argument("disable-gpu")
    options.add_argument("lang=ko_KR")
    #options.add_argument("headless")

    driver = webdriver.Chrome('C:\\Users\ ',options=options) # 크롬드라이버 경로
    driver.get(url)
    time.sleep(2)
    
    down = driver.find_element(by=By.XPATH, value='/html/body/div[3]/footer/div[1]/a[1]')
    ActionChains(driver).move_to_element(down).perform()
    time.sleep(2)

    for i in range(1,61):
        category.append(cat)
        
        k = 1
        while 1:
            if k<30:
                k = k+1
                try:                                                  
                    product11 = driver.find_element(by=By.XPATH, value='/html/body/div[3]/section/form/div/div/div[1]/div[%d]/ul/li[%d]/a/dl/dd/div[2]' %(k, i)).text
                    break
                except NoSuchElementException :
                    continue
            else:
                product11 = None
                break
        product1.append(product11)
        
        
        k = 1
        while 1:
            if k<30:
                k = k+1
                try:                                                  
                    price1 = driver.find_element(by=By.XPATH, value='/html/body/div[3]/section/form/div/div/div[1]/div[%d]/ul/li[%d]/a/dl/dd/div[3]/div/div[1]/em/strong' %(k, i)).text
                    break                                            
                except NoSuchElementException :                      
                    continue
            else:
                price1 = None
                break
        price.append(price1)
        
        
        k = 1
        while 1:
            if k<30:
                k = k+1
                try:                                                  
                    place1 = driver.find_element(by=By.XPATH, value='/html/body/div[3]/section/form/div/div/div[1]/div[%d]/ul/li[%d]/a/dl/dd/div[3]/div/div[1]/em/span/img' %(k, i)).get_attribute("alt")
                    break
                except NoSuchElementException :                     
                    continue
            else:
                place1 = '일반배송'
                break
        place.append(place1)
        
        
        k = 1
        while 1:
            if k<30:
                k = k+1
                try:                                                  
                    review11 = driver.find_element(by=By.XPATH, value='/html/body/div[3]/section/form/div/div/div[1]/div[%d]/ul/li[%d]/a/dl/dd/div[4]/div/span[2]' %(k, i)).text
                    break
                except NoSuchElementException :                     
                    continue
            else:
                review1 = None
                break
        review1.append(review11)
        
        
        k = 1
        while 1:
            if k<30:
                k = k+1
                try:                                                  
                    url1 = driver.find_element(by=By.XPATH, value='/html/body/div[3]/section/form/div/div/div[1]/div[%d]/ul/li[%d]/a' %(k, i)).get_attribute("href")
                    break
                except NoSuchElementException :
                    continue
            else:
                url1 = None
                break
        product_url.append(url1)

    time.sleep(1)
    driver.close()
```

## 리뷰 크롤링
#### 50개

```python
def review(cat, prod, url):
    
    options = webdriver.ChromeOptions()
    options.add_argument("disable-gpu")
    options.add_argument("lang=ko_KR")
    #options.add_argument("headless")

    driver = webdriver.Chrome('C:\\Users\ ',options=options) # 크롬드라이버 경로
    driver.get(url)
    time.sleep(2)
    
    body = driver.find_element_by_css_selector('body')
    
    for i in range(3):
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(random.random())
    
    try:
        driver.find_element(by=By.XPATH, value = '/html/body/div[2]/section/div[2]/div[10]/ul[1]/li[2]').click()
        time.sleep(2.5)

        for k in range(3, 13):

            for i in range(1,6):
                try:
                    id1 = driver.find_element(by=By.XPATH, value ='/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article[%d]/div[1]/div[2]/span' %i).text
                except:
                    id1 = None
                id2.append(id1)

                try:
                    rating1 = driver.find_element(by=By.XPATH, value = '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article[%d]/div[1]/div[3]/div[1]/div' %i).get_attribute('data-rating')
                except:
                    rating1 = None
                rating.append(rating1)

                try:
                    date1 = driver.find_element(by=By.XPATH, value = '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article[%d]/div[1]/div[3]/div[2]' %i).text
                except:
                    date1 = None
                date.append(date1)

                try:
                    main_review1 = driver.find_element(by=By.XPATH, value = '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article[%d]/div[4]/div' %i).text
                except:  
                    main_review1 = None
                main_review.append(main_review1)


                try:
                    like1 = driver.find_element(by=By.XPATH, value = '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article[%d]/div[6]' %i).get_attribute('data-count')
                except:                                              
                    j = 1
                    while 1:
                        if j<10:
                            j = j+1
                            try: 
                                like1 = driver.find_element(by=By.XPATH, value='/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/article[%d]/div[%d]/div[1]/strong/span' %(i, j)).text
                                break
                            except NoSuchElementException :
                                continue
                        else:
                            like1 = ''
                            break
                like.append(like1)

                category2.append(cat)
                product2.append(prod)
                
            try:
                next1 = driver.find_element(by=By.XPATH, value = '/html/body/div[2]/section/div[2]/div[10]/ul[2]/li[2]/div/div[6]/section[4]/div[3]/button[%d]' %k)
                ActionChains(driver).move_to_element(next1).click(next1).perform()
                time.sleep(random.uniform(0.7, 0.9))
            except:
                break
    
    except:
        pass
    
    driver.close()
```

```python

def result():
    base('https://www.coupang.com/')
    base_url = pd.DataFrame({'category':category, 'url':category_url})


    for i in range(len(base_url)):
        coupang_product(base_url.iloc[i,0], base_url.iloc[i,1])
        progress = round((i+1)/len(base_url),2)*100
        stdout.write("\r ===== %d%% completed =====" % progress)
    stdout.flush()
    stdout.write("\n")

    coupang_url = pd.DataFrame({'category':category1, 'product':product1, 'price':price, 'place':place, 'review':review1, 'url':product_url})

    coupang_url['쿠팡상품번호1'] = ''
    coupang_url['쿠팡상품번호2'] = ''

    for i in range(len(coupang_url)):
        try:
            coupang_url.iloc[i,4] = coupang_url.iloc[i,4].replace("(","").replace(")","")
        except:
            pass

        try :
            coupang_url['쿠팡상품번호1'][i] = coupang_url['url'][i].split('/')[5].split('?')[0]
            coupang_url['쿠팡상품번호2'][i] = coupang_url['url'][i].split('/')[5].split('&')[0].split('=')[1]
        except:
            coupang_url['쿠팡상품번호1'][i] = None
            coupang_url['쿠팡상품번호2'][i] = None


    category_list = list(coupang_url['category'].unique())
    print(category_list)
    while 1:
        select = str(input('카테고리 중 하나를 선택하세요.'))
        if select in category_list:
            coupang_url = coupang_url[coupang_url['category'] == select]
            break
        else:
            print('다시 선택해주세요.')
            continue


    for i in range(len(coupang_url)):
        coupang_review(coupang_url.iloc[i,0], coupang_url.iloc[i,1], coupang_url.iloc[i,5])
        progress = round((i+1)/len(coupang_url),2)*100
        stdout.write("\r ===== %d%% completed =====" % progress)
    stdout.flush()
    stdout.write("\n")
    
    global reviews
    reviews = pd.DataFrame({'category':category2, 'product':product2, 'user_id':id2, 'rating':rating, 'date':date, 'review': main_review, 'like': like})
    return reviews
```

```python
category = []
category_url = []

category1 = []
product1 = []
price = []
place = []
review1 = []
product_url = []

category2 = []
product2 = []
id2 = []
rating = []
date = []
main_review = []
like = []


result()
```
